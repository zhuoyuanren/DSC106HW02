```python
df = pd.read_csv('daily_sales.csv')
```

```python
df.head()
```

```python
df['total_sales'] = df.iloc[:,2:].sum(axis=1)
```

```python
df = df.groupby('Unnamed: 0')['total_sales'].mean().to_frame()
```

```python
df.reset_index(level=0, inplace=True)
```

```python
df = df.sort_values(by = ['total_sales'],ascending=False)
df
```

```python
df.to_csv('weekday_sale.csv',index=False)
```

```python
data = pd.read_csv('daily_sales.csv')
data['NE']=data['HM-NE']+data['CF-NE']+data['FF-NE']
data['SW']=data['HM-SW']+data['CF-SW']+data['FF-SW']
data['NW']=data['HM-NW']+data['CF-NW']+data['FF-NW']
data['SE']=data['HM-SE']+data['CF-SE']+data['FF-SE']
data['C']=data['HM-C']+data['CF-C']+data['FF-C']
```
